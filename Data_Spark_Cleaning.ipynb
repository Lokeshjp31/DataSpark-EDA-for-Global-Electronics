# Sales Table

```python
sales_df = pd.read_csv("Sales.csv")
sales_df
```

```python
#converting to datetime datatypes
sales_df['Order Date'] = pd.to_datetime(sales_df['Order Date'])
# converting to required format
sales_df['Order Date'] = sales_df['Order Date'].dt.strftime('%m/%d/%Y')
```

```python
sales_df['Order Date'] = pd.to_datetime(sales_df['Order Date'])
```

```python
print(sales_df.isnull().sum())
```

```python
# number of rows
total_rows=sales_df.shape[0]
```

```python
#percentage of null values
((sales_df.isnull().sum().sum())/total_rows)*100
```

```python
# 79% Delivery Date column is null filling with Not Available comment
sales_df.fillna('Not Available',inplace=True)
sales_df
```

```python
sales_df.dtypes
```

```python
sales_df.to_csv("Sales_df_new.csv",index=False)
```

# Store Table

# Over All Merge

```python
overall = sales_df
```

```python
overall=pd.merge(overall,customers_df,on='CustomerKey',how='inner')
```

```python
overall.dropna(inplace=True)
```

```python
overall
```

```python
overall=pd.merge(overall,exchange_rates_df,left_on=['Order Date','Currency Code'],right_on=['Date','Currency'],how='inner')
```

```python
overall=pd.merge(overall,products_df,on='ProductKey',how='inner')
```

```python
overall=pd.merge(overall,stores_df,on=['StoreKey','Country','State'],how='inner')
```

```python
overall
```

```python
col=list(overall.columns)
col1=[]
for c in col:
    col1.append(c.replace(' ','_').lower())
    
overall.columns=col1
```

```python
col=list(stores_df.columns)
col1=[]
for c in col:
    col1.append(c.replace(' ','_').lower())
    
stores_df.columns=col1
```

# Connecting to MySQL

```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine
```

```python
host = "localhost"
port = "3306"
username = "root"
password = "Root"
database = "Data_Spark"
```

```python
engine_string = f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}"
engine = create_engine(engine_string)
 
table_name = "overall"
overall.to_sql(table_name, engine,if_exists='replace',index=False)
print(f"successfully imported {table_name} to sql")

table_name = "customers"
customers_df.to_sql(table_name, engine,if_exists='replace',index=False)
print(f"successfully imported {table_name} to sql")

table_name = "products"
products_df.to_sql(table_name, engine,if_exists='replace',index=False)
print(f"successfully imported {table_name} to sql")

table_name = "exchange_rates"
exchange_rates_df.to_sql(table_name, engine,if_exists='replace',index=False)
print(f"successfully imported {table_name} to sql")

table_name = "stores"
stores_df.to_sql(table_name, engine,if_exists='replace',index=False)
print(f"successfully imported {table_name} to sql")

table_name = "sales"
sales_df.to_sql(table_name, engine,if_exists='replace',index=False)
print(f"successfully imported {table_name} to sql")
```
